```python
from ConversionUtilities import *
from matplotlib import pyplot as plt
from IPython.display import Audio
import numpy as np
```

```python
f = open("VoicemailRaw", "rb")
rawData = bytearray(f.readlines()[0])
intensity = getAudacityFormat(rawToHexString((bytes(rawData))).decode()) 
plt.plot(np.arange(0, len(intensity))/8000, intensity)
plt.title("Original Signal")
plt.show()
Audio(np.array(intensity), rate=8000)
```

```python
message = "Hello from the raw audio!"

bitCounter = 0
k = 0
while bitCounter < len(message) << 4:
    rawData[k] = rawData[k] | 1 if (ord(message[bitCounter >> 4]) >> (bitCounter & 7)) & 1 else rawData[k] & ~1
    bitCounter += 1
    k += 1
intensity2 = getAudacityFormat(rawToHexString((bytes(rawData))).decode()) 

print(rawData)

plt.plot(np.arange(0, len(intensity2))/8000, intensity2)
plt.title("Encoded Signal")
plt.show()

Audio(np.array(intensity2), rate=8000)
```

```python
plt.plot(np.arange(0, len(intensity2))/8000, np.array(intensity) - np.array(intensity2))
plt.title("Diff Signal")
plt.show()
```
