```python
import binascii
import numpy as np
import matplotlib
from rawConverter import *
from matplotlib import pyplot as plt
from IPython.display import Audio
import audioop
```

```python
file = open("VoicemailASCIIHex", "r")  # Original voicemail file
lines = file.readlines()

originalIntensity = []
for line in lines:
    originalIntensity += getAudacityFormat(line)   

print(lines)

plt.plot(originalIntensity)
plt.title("Original Signal")
plt.show()

Audio(np.array(originalIntensity), rate=8000, autoplay=True)
    
```

```python
message = "Hello. I am communicating to you through the signal!"
if len(message) % 2 != 0:
    message += " "

lines = encodeMessagesInAudioPayloads(lines, messageToAsciiHexString(message))
intensity = []
for line in lines:
    intensity += getAudacityFormat(line)     
    
plt.plot(intensity)
plt.title("Encoded Signal")
plt.show()

Audio(np.array(intensity), rate=8000, autoplay=True)

```

```python
plt.plot(np.array(originalIntensity) - np.array(intensity))
plt.title("Error Signal")
plt.show()

Audio(np.array(originalIntensity) - np.array(intensity), rate=8000, autoplay=True)
```
